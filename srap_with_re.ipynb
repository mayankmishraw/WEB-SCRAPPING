```python
url = "https://www.imdb.com/chart/top/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
tag=soup.find('div', class_=re.compile(r'by+'))
tag
```
